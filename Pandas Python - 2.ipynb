# Treating missing values in Python

```python
cars.loc[5,'hp'] = np.nan
cars.loc[3,'hp'] = np.nan
cars.loc[6,'hp'] = np.nan
cars.loc[5,'mpg'] = np.nan
cars.loc[2,'mpg'] = np.nan
```

```python
cars.head(10)
```

```python
cars.isnull().sum()
```

```python
cars.fillna(100)
```

```python
# Treating missing values with forward filling. Filling with value abobe the missing value
cars.fillna(method="ffill")
```

```python
# Treating missing values with backward filling. Filling with value abobe the missing value
cars.fillna(method="bfill")
```

```python
# Treating missing values with mean value. 
cars.fillna(cars.mean())
```

# Feature Creation

```python
# Create a new column : Bangalore , 0/1 value condition is with cars > 20 mpg = 1: cars <= 20 mpg = 0
bang = DataFrame(np.ceil(np.random.rand(33)))
bang.columns = ['Banglaore']
bang
```

```python
cars.shape
```

```python
bang.head()
```

```python
cars.head()
```

```python
cars.loc[cars.mpg>20, 'Bangalore'] = 1
cars.loc[cars.mpg<=20, 'Bangalore'] = 0
cars.head()
```

```python
# Create a  new feature light cars with wt >3 as 0 and lesser &= 1
```

```python
cars.loc[cars.wt>3, 'Light Cars'] = 0
cars.loc[cars.wt<=3, 'Light Cars'] = 1
cars.head()
```
